```python
def showSegmentedImages(masks, segmentedImages):
    nSegmented = len(masks)
    figureHeight = 5*nSegmented
    fig, ax = plt.subplots(nSegmented, 1, figsize = (5, figureHeight))

    if(nSegmented == 1):
        ax.imshow(segmentedImages[0])
        ax.set_axis_off()

    elif(nSegmented > 1):
        for i in range(0,nSegmented):
            ax[i].imshow(segmentedImages[i])
            ax[i].set_axis_off()
```

```python
from SellerStandSegmentation import SellerStandSegmentator


segmentator = SellerStandSegmentator("stand_detector_105_iters.pt")

masks, segmentedImages = segmentator.segmentSellerStands(imgTest)

showSegmentedImages(masks, segmentedImages)

```
